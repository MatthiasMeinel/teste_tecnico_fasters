```python
#Criando SparkSession
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.appName("AnaliseExploratoria").getOrCreate()
```

```python
#Lendo os arquivos CSV
```

```python
avaliacoes = spark.read.csv("base_avaliacoes.csv", header=True, inferSchema=True)
pessoas = spark.read.csv("base_pessoas.csv", header=True, inferSchema=True)
telefonia = spark.read.csv("base_telefonia.csv", header=True, inferSchema=True)
```

```python
avaliacoes.show()
```

```python
pessoas.show()
```

```python
telefonia.show()
```

```python
#Para achar primeiro uma tabela de vendas, é necessário filtrar a tabela telefonia
```

```python
vendas = telefonia.filter(col("Motivo")=="Venda")
```

```python
#Agora temos uma tabela apenas de vendas
```

```python
vendas.show()
```

```python
#Agora para acharmos o top 5 vendedores precisaremos fazer algumas agregações na tabela
```

```python
top_vendedores = vendas.groupBy("Username").agg(sum("Valor venda"))
```

```python
top_vendedores.show()
```

```python
#É importante ordenar e limitar o top 5 para facilitar visualização
```

```python
top_vendedores = vendas.groupBy("Username").agg(sum("Valor venda").alias("Total Vendas")).orderBy(col("Total Vendas").desc()).limit(5)
```

```python
top_vendedores.show()
```

```python
#Agora para acharmos o nome a partir do valor de username precisamos fazer um join com a tabela de pessoas
```

```python
top5=top_vendedores.join(pessoas, "Username", "left").select("Username", "Nome", "Total Vendas")
```

```python
top5.show()
```

```python
#Agora vamos achar o ticket médio por vendedor
```

```python
from pyspark.sql.functions import avg

ticket_medio = vendas.groupBy("Username").agg(avg("Valor venda").alias("ticket_medio"))
```

```python
ticket_medio.show()
```

```python
#Para adicionar o nome dos vendedores
```

```python
ticket_medio=ticket_medio.join(pessoas, "Username", "left").select("Username", "Nome", "ticket_medio")
```

```python
ticket_medio.show()
```

```python
#Agora para achar tempo médio de ligações
```

```python
telefonia.show()
```

```python
telefonia = telefonia.withColumn("inicio_ts", col("inicio_ligacao").cast("timestamp")).withColumn("fim_ts", col("fim_ligação").cast("timestamp"))
```

```python
telefonia = telefonia.withColumn("duracao_segundos", unix_timestamp("fim_ts") - unix_timestamp("inicio_ts"))
```

```python
#com isso, podemos calcular o tempo médio em segundos
```

```python
tempo_medio = telefonia.agg(avg("duracao_segundos").alias("tempo_medio"))
```

```python
tempo_medio.show()
```

```python
telefonia.filter(col("duracao_segundos") < 0).show()
```

```python
#Foi achado uma inconsistência nos dados - 2 valores deram menor que 0
```

```python
#Vou filtar apenas os dados consistentes e então refazer o cálculo
```

```python
telefonia_validas = telefonia.filter(col("duracao_segundos") >= 0)
```

```python
tempo_medio = telefonia_validas.agg(
    round(avg("duracao_segundos")/60, 2).alias("tempo_medio_minutos")
)

tempo_medio.show()
```

```python
#Dessa forma o tempo médio das ligações com dados válidos foi de 9.92 minutos
```

```python
#Agora para a análise das notas
```

```python
nota_media = avaliacoes.agg(avg("Nota").alias("nota_media_geral"))
nota_media.show()
```

```python
avaliacoes_join = avaliacoes.join(pessoas, "Username", "left")

pior_vendedor = avaliacoes_join.groupBy("Username", "Nome").agg(avg("Nota").alias("nota_media")).orderBy("nota_media").limit(5)

pior_vendedor.show()
```
